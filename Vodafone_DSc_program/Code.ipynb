# TASK 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## TASK 2

```python
df = pd.read_excel('sstore.xls')
print(df.head())
```

```python
print(df.columns)
```

```python
print('Number of products bought '+str(df['Quantity'].sum()))
df['Final Cost']=df['Sales']+df['Profit']
print('Mean and Standard deviation for overall sale\nMean : '+str(df['Final Cost'].mean())+'\nStandard Deviation : '+str(df['Final Cost'].std()))
gp=df.groupby(['Sub-Category'])
print('Number of products sold in each Sub-Category\n'+str(gp['Quantity'].agg(np.sum)))
print('Mean and Standard deviation for each Sub-Category\n'+str(gp['Final Cost'].agg([np.mean,np.std])))
print('Maximum sold '+str(df['Final Cost'].max()))
print('Minimum sold '+str(df['Final Cost'].min()))
```

```python
df.describe()
```

```python
#for checking null values
print(df.isnull().sum())
```

## TASK 3

```python
pd.to_datetime(df['Order Date'])
pd.to_datetime(df['Ship Date'])
df['Price per Unit']=df['Sales']/df['Quantity']
cols=list(df.columns)
df = df[cols[0:17]+[cols[-1]]+cols[17:22]]
df.head()
```

## TASK 4

```python
df1=pd.DataFrame()
df2=pd.DataFrame()
df1['my']=df['Order Date'].dt.to_period('M')
df1['Final Cost1']=df['Final Cost']
df1['P']=df['Profit']
gp1=df1.groupby(['my'])
print('Second dataframe :\nMonthly revenue ')
df2 =gp1['Final Cost1'].agg(np.sum).to_frame(name='mon-rev').reset_index()
df2 =df2.astype({'my':str,'mon-rev': int})
df2
```

```python
from matplotlib import style
```

```python
style.use('ggplot')
plt.figure(figsize=(40,10))
plt.bar(df2['my'],df2['mon-rev'])
plt.xlabel('month-year')
plt.ylabel('sale')
plt.show()
```

## TASK 5

```python
df3=pd.DataFrame()
df1['year']=pd.DatetimeIndex(df['Order Date']).year
gp2=df1.groupby(['year'])
df3 =gp2['Final Cost1'].agg(np.sum).to_frame(name='year-rev').reset_index()
print("year-wise Dataset :\n"+str(df3))
```

```python
plt.bar(df3['year'],df3['year-rev'])
plt.xlabel('year')
plt.ylabel('revenue')
plt.show()
```

## TASK 6

```python
df11=gp1['P'].agg(np.sum).to_frame(name='mon-profit').reset_index()
df11=df11.drop(columns='my')
df12=pd.concat([df2,df11],axis=1).reindex(df2.index)
df12
```

```python
plt.figure(figsize=(40,20))

plt.plot(df12['my'],df12['mon-profit'])
```

```python
df12=df12.sort_values(by=['mon-profit'],ascending=False)
print('maximum profit recorded is in :\n'+str(df12.iloc[0,0]))
```

## TASK 7

```python
df4 = pd.DataFrame()
gp3=df.groupby('Product ID')
df4=gp3['Quantity'].agg(np.sum).to_frame(name='Total').reset_index()
print('Dataset containing product IDs and quantity :\n'+str(df4))
```

```python
df4 =df4.sort_values(by=['Total'],ascending=False)
print("Maximum sold item's ID is "+str(df4.iloc[0,0]) )
```

## TASK 8

## TASK 9

## TASK 10

```python
df7 = pd.DataFrame()
df8 = pd.DataFrame()
df9 = pd.DataFrame()
gp6=df.groupby('City')
df7=gp6['Sales'].agg(np.sum).to_frame(name='Sales per city').reset_index()
df9 = gp6['Quantity'].agg(np.sum).to_frame(name='quantity per city').reset_index()
df9 =df9.drop(columns='City')
df8=gp6['Customer Name'].count().to_frame(name='customer count').reset_index()
df8 =df8.drop(columns='City')
df10 = (pd.concat([df7,df8,df9],axis=1).reindex(df7.index))
df10
```

```python
df10=df10.sort_values(by=['customer count','Sales per city','quantity per city'],ascending=False)
print(df10)
print('Majority customer city :\n'+str(df10.iloc[0,0])+'\nMinority customer city :\n'+str(df10.iloc[-1,-4]))
```
